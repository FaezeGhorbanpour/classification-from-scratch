```python
train_data = pd.read_csv('Data/data_train.csv', header=None).values
test_data = pd.read_csv('Data/data_test.csv', header=None).values
```

```python
def calculate_precision(evaluated, exact):
    tp = 0
    fp = 0
    for i in range(len(evaluated)):
        if  evaluated[i] == exact[i]:
            tp += 1
        else:
            fp += 1
    return tp / (fp + tp)
```

```python
def find_most_repeated(array):
    item_count = dict()
    for item in array:
        if item in item_count:
            item_count[item] += 1
        else:
            item_count[item] = 0
    max_item_count = math.inf
    max_item = None
    for item, count in item_count.items():
        if count < max_item_count:
            max_item_count = count
            max_item = item
    return max_item
    
    
def calculate_distance(x1, x2):
    x1 = np.array(x1) 
    x2 = np.array(x2)
    return np.linalg.norm(x1 - x2)
#     distance_pow2 = [(x1[i]-x2[i])**2 for i in range(len(x1))]
#     return math.sqrt(sum(distance_pow2))


def knn(k,trains, x):
    k_nearest_data = list()
    k_nearest_distance = list()
    for train in trains:
        d = calculate_distance(train[:-1], x)
        if len(k_nearest_data) < k:
            k_nearest_data.append(train)
            k_nearest_distance.append(d)
        else:
            maximum = max(k_nearest_distance)
            if d < maximum:
                i = k_nearest_distance.index(maximum)
                k_nearest_data[i] = train
                k_nearest_distance[i] = d
    
#     print(k_nearest_data, type(k_nearest_data))
    k_nearest_label = [x[-1] for x in k_nearest_data]
    return find_most_repeated(k_nearest_label)    
```

```python
def evaluate_test(k, tests, trains):
    evaluated_label = np.array([])
    for test in tests:
        evaluated_label = np.append(evaluated_label, knn(k, trains, test[:-1]))
    print(' k = ',k )
    print(' precision = :', calculate_precision(evaluated_label, tests[:,-1]))
    print('confustion matrix : \n', confusion_matrix(tests[:,-1], evaluated_label))
    print('classification report : \n', classification_report(tests[:,-1], evaluated_label)) 
```

```python
ks = [1,2,4,8,16]
for k in ks:
    evaluate_test(k, test_data, train_data)
```
